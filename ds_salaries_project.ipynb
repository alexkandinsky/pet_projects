##<b>Описание данных</b>

- <b>work_year</b> - год получения зарплаты;
- <b>experience_level</b> - опыт работы (EN - Junior, MI - Middle, SE - Senior, EX - Director);
- <b>employment_type</b> - тип найма (FT - полный, PT - частичный, CT - контракт, FL - фриланс);
- <b>job_title</b> - наименование должности;
- <b>salary</b> - зарплата;
- <b>salary_currency</b> - валюта заработной платы;
- <b>salary_in_usd</b> - зароботная плата в долларах;
- <b>employee_residence</b> - место работы сотрудника;
- <b>remote_ratio</b> - удаленная работа (0 - нет, 50 - частично, 100 - полностью);
- <b>company_location</b> - место расположения компании;
- <b>company_size</b> - размер компании (S - меньше 50, M - от 50 до 250, L - 250 + сотрудников).

###Распределение работы в зависимости от года
