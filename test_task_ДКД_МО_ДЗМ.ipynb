```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Тестовое_задание_ДКД_МО_ДЗМ_Аналитик (2).xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Дата начала заболевания'] = df['Дата начала заболевания'].dt.strftime('%Y-%m-%d')
```

```python
df['Дата выздоровления'] = df['Дата выздоровления'].dt.strftime('%Y-%m-%d')
```

```python
df['Тяжесть'].fillna('Нет информации', inplace=True)
```

```python
print('Общее кол-во пациентов:', df['ID пациента'].nunique())
```

```python
df['ID пациента'].duplicated().sum()
```

```python
print('Общее кол-во заболеваний:', df['ID заболевания'].count())
```

```python
df['ID заболевания'].duplicated().sum()
```

```python
count_by_date_disease = df.groupby('Дата начала заболевания')['ID заболевания'].count()
```

```python
plt.figure(figsize=(14, 6))
count_by_date_disease.plot(kind='bar', color='lightgreen')
plt.xlabel('Дата начала заболевания')
plt.ylabel('Кол-во болезней')
plt.title('Ежедневная динамика прироста инфекции')
plt.show()
```

```python
count_by_date_disease.reset_index().
```

```python
count_by_date_disease.describe()
```

```python
age_in_days = pd.to_datetime('today') - df['ДР']
```

```python
df['Возраст пациента'] = (age_in_days.dt.days.astype(float) / 365).round(1)
```

```python
df['Возрастная группа'] = pd.cut(df['Возраст пациента'], bins=[0, 17, 64, float('inf')], labels=['0-17', '18-64', '65+'], right=False)
```

```python
df.head()
```

```python
df.groupby(['Пол', 'Возрастная группа',  'Тяжесть'])['ID заболевания'].count().unstack()
```

```python
medical_organizations = ['Поликлиника №1', 'Поликлиника №2', 'Поликлиника №3', 'Поликлиника №4', 'Поликлиника №5']
attached_patients = [7000, 17000, 22000, 9000, 11000]
```

```python
for med_org, patients in zip(medical_organizations, attached_patients):
   df.loc[df['Медицинская организация'] == med_org, 'Прикрепленное население'] = patients
```

```python
df['Прикрепленное население'] = df['Прикрепленное население'].astype(int)
```

```python
df.head()
```

```python
attached_and_disease = df.groupby(['Медицинская организация', 'Прикрепленное население'])['ID заболевания'].count().reset_index().rename(columns={'ID заболевания':'Кол-во заболевших пациентов'})
```

```python
attached_and_disease['Соотношение кол-ва заболевших пациентов / прикрепленного населения'] = attached_and_disease['Кол-во заболевших пациентов'] / attached_and_disease['Прикрепленное население']
```

```python
attached_and_disease['Соотношение кол-ва заболевших пациентов / прикрепленного населения'] = attached_and_disease['Соотношение кол-ва заболевших пациентов / прикрепленного населения'].round(2) * 100
```

```python
attached_and_disease.sort_values(by='Соотношение кол-ва заболевших пациентов / прикрепленного населения', ascending=False)
```

```python
df.head()
```

```python
df['Дата начала заболевания'] = pd.to_datetime(df['Дата начала заболевания'])
```

```python
df['Продолжительность болезни'] = df['Дата выздоровления'] - df['Дата начала заболевания']
```

```python
df['Продолжительность болезни'].mean().days
```

```python
mean_duration_disease_by_group = df.groupby(['Пол', 'Возрастная группа', 'Тяжесть'])['Продолжительность болезни'].mean()
```

```python
mean_duration_disease_by_group.apply(lambda x: x.days).reset_index()
```
